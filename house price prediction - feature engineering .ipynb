# Importing training and test data



```python
# Training data
dataset = pd.read_csv('train.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
# Test data
predict_data = pd.read_csv('test.csv')
predict_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Handle missing values

First we will take care of the training set and then the test set.

## Training set

```python
# first we will take care of categorical variables
cat_nan_train = [feature for feature in X_train.columns if X_train[feature].isnull().sum()>1 and X_train[feature].dtypes=='O']

for feature in cat_nan_train:
    print(feature,np.round(X_train[feature].isnull().mean(),4),'% missing values')
```

```python
# replace missing values with a new label , a new category
def replace_cat_feature(dataset,features_nan):
    data = dataset.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data

X_train = replace_cat_feature(X_train,cat_nan_train)
X_train[cat_nan_train].isnull().sum()
```

```python
# now we see numerical features

numeric_nan_train = [feature for feature in X_train.columns if X_train[feature].isnull().sum()>1 and X_train[feature].dtypes!='O']

for feature in numeric_nan_train:
    print("{}: {} % missing values".format(feature, np.round(X_train[feature].isnull().mean(),4)))
```

```python
# since the numerical features had a lot of outliers, we are replacing the missing values using the median values

for feature in numeric_nan_train:
    median_value = X_train[feature].median()
    
    # create a new feature that stores which entry had a missing value
    X_train[feature+'nan'] = np.where(X_train[feature].isnull(),1,0)
    X_train[feature].fillna(median_value,inplace=True)
    
X_train[numeric_nan_train].isnull().sum()
```

```python
# now we see the temporal variables
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    X_train[feature] = X_train['YrSold'] - X_train[feature]
    
X_train[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

## Test set

```python
cat_nan_test = [feature for feature in X_test.columns if X_test[feature].isnull().sum()>1 and X_test[feature].dtypes=='O']

for feature in cat_nan_test:
    print(feature,np.round(X_test[feature].isnull().mean(),4),'% missing values')
```

```python
X_test = replace_cat_feature(X_test,cat_nan_test)
X_test[cat_nan_test].isnull().sum()
```

```python
numeric_nan_test = [feature for feature in X_test.columns if X_test[feature].isnull().sum()>1 and X_test[feature].dtypes!='O']

for feature in numeric_nan_test:
    print("{}: {} % missing values".format(feature, np.round(X_test[feature].isnull().mean(),4)))
```

```python
for feature in numeric_nan_test:
    median_value = X_test[feature].median()
    
    # create a new feature that stores which entry had a missing value
    X_test[feature+'nan'] = np.where(X_test[feature].isnull(),1,0)
    X_test[feature].fillna(median_value,inplace=True)
    
X_test[numeric_nan_test].isnull().sum()
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    X_test[feature] = X_test['YrSold'] - X_test[feature]
    
X_test[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

# Numerical Variables

We observed that the numerical variables are skewed, hence we will perform a log normal distribution

## Training set

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']

for feature in num_features:
    X_train[feature] = np.log(X_train[feature])
y_train=np.log(y_train)
X_train.head()    
```

```python
y_train.head()
```

## Test set

```python
for feature in num_features:
    X_test[feature] = np.log(X_test[feature])
y_test = np.log(y_test)
X_test.head()    
```

```python
y_test.head()
```

# Handling Rare Categorical Feature
We will remove categorical features that are present less than 1% of the observations

```python
categorical_features = [feature for feature in X_train.columns if X_train[feature].dtypes=='O']
categorical_features
```

```python
X_train['SalePrice']=y_train
for feature in categorical_features:
    temp = X_train.groupby(feature)['SalePrice'].count()/len(X_train)
    temp_df = temp[temp>0.01].index
    X_train[feature] = np.where(X_train[feature].isin(temp_df),X_train[feature],'Rare_var')
    X_test[feature] = np.where(X_test[feature].isin(temp_df),X_test[feature],'Rare_var')
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
for feature in categorical_features:
    labels_ordered=X_train.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    X_train[feature]=X_train[feature].map(labels_ordered)
    X_test[feature]=X_test[feature].map(labels_ordered)
```

## Feature Scaling

```python
feature_scale = [feature for feature in X_train.columns if feature not in ['Id','SalePrice']]

# StandardScaler converts data based on normal distribution
# MinMaxScaler converts data between 0 to 1 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train[feature_scale])
#scaler.transform(X_test[feature_scale])
```

```python
X_train_updated = pd.concat([X_train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_train[feature_scale]), columns=feature_scale)],
                    axis=1)
```
