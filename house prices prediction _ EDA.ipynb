# DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('train.csv')

print(dataset.shape)
```

```python
dataset.head()
```

## 1. Missing values

```python
features_with_null = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
```

```python
print('Features with Missing Values')
print('\n')
for feature in features_with_null :
    print(feature,np.round(dataset[feature].isnull().mean(),4) ,'%')
```

### Relationship between missing values and the Sales Price

```python
for feature in features_with_null:
    data=dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

From the above plots, it can be found that for some features like 'MasVnrType','LotFrontage', the SalesPrice is high whenever there are null values. Hence there is a relationship among the missing values and the SalesPrice

## 2. Numerical variables

```python
numeric_variables = [feature for feature in dataset.columns if dataset[feature].dtypes!='O']

print('Number of numerical variables :' , len(numeric_variables))

dataset[numeric_variables].head()
```

#### Temporal variables (Date Time variables)

```python
year_feature = [feature for feature in dataset.columns if 'Yr' in feature or 'Year' in feature]

print(year_feature)
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# lets see if there is any relationship between the year a house is sold and the Sale price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Sale price')
plt.title('House price vs Year Sold')
plt.show()
```

```python
# lets see if there is any relationship between the year a house is build and the Sale price

dataset.groupby('YearBuilt')['SalePrice'].median().plot()
plt.xlabel('Year built')
plt.ylabel('Sale price')
plt.title('House price vs Year Built')
plt.show()
```

```python
# lets see if there is any relationship between the year a house is remodelled and the Sale price

dataset.groupby('YearRemodAdd')['SalePrice'].median().plot()
plt.xlabel('Year house was remodelled')
plt.ylabel('Sale price')
plt.title('House price vs Year Remodelled')
plt.show()
```

```python
# lets see if there is any relationship between the year a house got its garage built and the Sale price

dataset.groupby('GarageYrBlt')['SalePrice'].median().plot()
plt.xlabel('Year garage was built')
plt.ylabel('Sale price')
plt.title('House price vs Year garage built')
plt.show()
```

```python
year_feature
```

```python
# We will compare the difference of all years with sales price

for feature in year_feature:
    if feature!='YrSold':
        data = dataset.copy()
        
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

It can be seen that as the house becomes old, the sale price decreases.

```python
# Numerical variables are of two types:
# Discrete and Continuous variables

discrete_feature = [feature for feature in numeric_variables if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]

print('Number of discrete features:',len(discrete_feature))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# Relationship between them and sales price

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature = [feature for feature in numeric_variables if feature not in discrete_feature+year_feature+['Id']]
print('Continuous feature count', len(continuous_feature))
```

```python
# we'll see the distribution of the continuous variables using histograms

for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

Some variables are skewed distributions. They are not normally distributed. We will be using logarithmic transformation.

```python
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

## 3. Outliers

```python
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        # detect outliers using boxplot
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
