```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns',None)
```

```python
X_train = pd.read_csv('X_train_updated.csv')
X_train.head()
```

```python
y_train = X_train[['SalePrice']]

X_train = X_train.drop(['Id','SalePrice'],axis = 1)
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_feat = X_train.columns[(feature_sel_model.get_support())]

print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_==0)))
```

```python
selected_feat
```

```python
X_train = X_train[selected_feat]
```

```python
X_train.to_csv('X_train_final.csv',index=False)
y_train.to_csv('y_train_final.csv',index=False)
```

## Test data

```python
X_test = pd.read_csv('X_test_updated.csv')
X_test.head()
```

```python
X_test = X_test[selected_feat]
```
